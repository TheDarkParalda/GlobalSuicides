# Exploratory Data Analysis on Global Suicides Notebook


The dataset for our project is a combined from several datasets made by United Nations Development Program, World Bank and World Health Organization. The dataset can be found here.The dataset has records from 1985 to 2016. However, since there is very few data in 2016, we will only keep the range from 1985 to 2015.The dataset has 27820 observations and 8 features.<br><br>

The features are : <br> &emsp;&emsp;
    - Country<br> &emsp;&emsp;
    - Year : from 1985 to 2015<br> &emsp;&emsp;
    - Sex<br> &emsp;&emsp;
    - Age : Age groups including “5-14”, “15-24”, “25-34”, “35-54”, “55-74”, and “75+”.<br> &emsp;&emsp;
    - Suicides_no : Number of Suicides<br> &emsp;&emsp;
    - Population<br> &emsp;&emsp;
    - GDP_per_capita : Ratio between the country’s GDP and it's Population<br> &emsp;&emsp;
    - suicides/100k pop<br> &emsp;&emsp;<br>

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("suicideset.csv")
print(data)
print(len(data))
```

## Analysis on Sex

Now let's see how sex affects the suicides and see it's trend through the years.

```python
Sp100k_Male = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['sex']=='male'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]
Sp100k_Female = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['sex']=='female'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]

```

```python
Mean_Sp100k_Male = np.mean(Sp100k_Male)
Mean_Sp100k_Female = np.mean(Sp100k_Female)
print("Average Suicides per 100k Man : ", Mean_Sp100k_Male)
print("Average Suicides per 100k Woman : ", Mean_Sp100k_Female)
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
points = [1987 + year for year in range(2016-1987)]

ax.plot(points, (Sp100k_Male), linewidth=1.0,marker = "o")
ax.plot(points, (Sp100k_Female), linewidth=1.0,marker = "o",color = "tab:pink")

plt.legend(['Male Suicides per 100k' , 'Female Suicides per 100k'],loc = 'upper right')
```

```python
barWidth = 0.4
fig = plt.subplots(figsize =(12, 8))

br1 = np.arange(len(Sp100k_Male))
br2 = [x + barWidth for x in br1]

plt.bar(br1, Sp100k_Male, color ='tab:blue', width = barWidth,
        edgecolor ='grey', label ='Male')
plt.bar(br2, Sp100k_Female, color ='tab:pink', width = barWidth,
        edgecolor ='grey', label ='Female')

plt.legend(['Male Suicides per 100k' , 'Female Suicides per 100k'],loc = 'upper right')
```

From this graph it is clear that a man has a more likely probability of suicide than a female. On average a man has 4 times more likelyhood of suicide than a woman.

## Analysis on Age Groups

Now let's see how age group affects the suicides and see it's trend through the years.

```python
Sp100k_5 = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['age']=='5-14 years'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]
Sp100k_15 = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['age']=='15-24 years'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]
Sp100k_25 = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['age']=='25-34 years'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]
Sp100k_35 = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['age']=='35-54 years'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]
Sp100k_55 = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['age']=='55-74 years'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]
Sp100k_75 = [float(np.squeeze((data.loc[(data['year']==1987 + year)&(data['age']=='75+ years'),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32))) for year in range(2016-1987)]


Mean_Sp100k_5 = np.mean(Sp100k_5)
Mean_Sp100k_15 = np.mean(Sp100k_15)
Mean_Sp100k_25 = np.mean(Sp100k_25)
Mean_Sp100k_35 = np.mean(Sp100k_35)
Mean_Sp100k_55 = np.mean(Sp100k_55)
Mean_Sp100k_75 = np.mean(Sp100k_75)
print("Average Suicides per 100k 5-14 years : ", Mean_Sp100k_5)
print("Average Suicides per 100k 15-24 years : ", Mean_Sp100k_15)
print("Average Suicides per 100k 25-34 years : ", Mean_Sp100k_25)
print("Average Suicides per 100k 35-54 years : ", Mean_Sp100k_35)
print("Average Suicides per 100k 55-74 years : ", Mean_Sp100k_55)
print("Average Suicides per 100k 75+ years : ", Mean_Sp100k_75)
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
points = [1987 + year for year in range(2016-1987)]

ax.plot(points, (Sp100k_5), linewidth=1.0,marker = "x")
ax.plot(points, (Sp100k_15), linewidth=1.0,marker = "x")
ax.plot(points, (Sp100k_25), linewidth=1.0,marker = "x")
ax.plot(points, (Sp100k_35), linewidth=1.0,marker = "x")
ax.plot(points, (Sp100k_55), linewidth=1.0,marker = "x")
ax.plot(points, (Sp100k_75), linewidth=1.0,marker = "x")

plt.legend(['Suicides per 100k 5-14 years' , 'Suicides per 100k 15-24 years' , 'Suicides per 100k 25-34 years', 'Suicides per 100k 35-54 years' , 'Suicides per 100k 55-74 year', 'Suicides per 100k 75+ years'],loc = 'lower right', bbox_to_anchor =(1.7, 0))

```

From this it can be observed that as people get older, more they have the tendency to commit suicides.

```python
fig = plt.subplots(figsize =(12, 8))

plt.bar(['5-14 years' , '15-24 years' , '25-34 years', '35-54 years' , '55-74 year', '75+ years'], [Mean_Sp100k_5,Mean_Sp100k_15,Mean_Sp100k_25,Mean_Sp100k_35,Mean_Sp100k_55,Mean_Sp100k_75], edgecolor ='grey')

plt.legend(['Suicides per 100k 5-14 years' , 'Suicides per 100k 15-24 years' , 'Suicides per 100k 25-34 years', 'Suicides per 100k 35-54 years' , 'Suicides per 100k 55-74 year', 'Suicides per 100k 75+ years'],loc = 'upper right')
```

## Countries and Suicides

Now we change our perspective to continent and find the suicide trend according to each continents and also how the age group and continent are related.

```python
Sp100k_country = [float((data.loc[(data['country']==c),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32)) for c in data["country"].drop_duplicates()]
```

```python
#print(Sp100k_country)
```

```python
fig = plt.subplots(figsize =(12, 20))

plt.barh([c for c in data["country"].drop_duplicates()], Sp100k_country, edgecolor ='grey')
```

From the above barplot it is evident that Lithuvania has the highest suicides per 100k in the world.

As a college student myself, I want to know which country has the most suicide rate for 15-24 years category.

```python
Sp100k_countryC = [float((data.loc[((data['country']==c)&(data['age']=='15-24 years')),["suicides/100k pop"]].mean()).to_numpy().astype(np.float32)) for c in data["country"].drop_duplicates()]

print(Sp100k_countryC)
```

```python
fig = plt.subplots(figsize =(12, 20))

plt.barh([c for c in data["country"].drop_duplicates()], Sp100k_countryC, edgecolor ='grey',color = 'tab:orange')
```

Sri Lanka has the most number of college student Suicides.
